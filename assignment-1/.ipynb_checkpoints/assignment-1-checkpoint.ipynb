<h1>Exercise 1</h1>
<p>
Write a function that takes as argument a list of floats and calculates the sum of the numbers in the list by just using basic operations. The function should return a float or null in case the list is empty.    
</p>

<div>
    <h3><b>Logic <b></h3>
    <p>
    <ul>
        <li>
            The function is restricted to use float values as input. 
        </li>
        <li>
            The output is restritcted to retrun Float or None with a <strong>Union</strong> of the types
        </li>
         <li>
            When the list is empty <strong>None</strong> is returned
        </li>
         <li>
             All valid items in the list are added before return
        </li>
    </ul>
</p>
</div>

<div>
    <h3><b>Please Note <b></h3>
<ol>
    <li>
        Due to the nature of jupyterbook strict types/typechecking at function declaration are not working as expected unlike VS code IDE.
    </li>
    <li> 
        For an array passed as a list of mix type objects ( float, strings). I had to put an extra check <br\>
        </br><strong>if isinstance(number ,(float,int)):</strong> to find valid numbers
    </li>
</ol>
</div>


```python
from typing import Union, List

def addition(numbers: List[float]) -> Union[float,None]:
    sum = 0.0
    if numbers is None or len(numbers) ==0:
       return None
    else: 
        for number in numbers:
            if isinstance(number ,(float,int)):
                sum = sum+ number
        return sum
```

<h3>Test cases</h3>
<h4>When an empty list is passed</h4>

```python
print(addition([]))
```

2. When floating list is passed

```python
print(addition([1.8,2.7,3.8,4.4,5.13]))
```

3. When integers are passed

```python
print(addition([1,2,3,4,5]))
```

4. when combination of string and numbers are passed

```python
print(addition([1,'2',3,'4',5]))
```

<h1>Exercise 2</h1>
<p>
Palindromes are sequences of characters that read the same forward and backward. Write a function that takes a string as an argument and returns a boolean flag that is true if the string is a palindrome and false otherwise.    
</p>

<div>
    <h3><b>Logic <b></h3>
    <p>
    <ul>
        <li>
            If input is <strong>None</strong> is_palindorome method returns <strong>None</strong>
        </li>
        <li>
            In place of reversing the complete string and comparing with the original to test palindorme
            The string is divided into two equal parts, the second part is reverrsed and compared to the first part.
            In place of storing parts in variables the comparison in If and else could be done, but to write better clear code this step was not intentionally done.
        </li>
    </ul>
</p>
</div>

<div>
    <h3><b>Please Note <b></h3>
            <p>
                Since a <strong>uppercase</strong> character and <strong>smallcase</strong> character have clear distinction in terms of having different <strong>ordinals</strong>, this program treats 'A' and 'a' as different characters, if needed this behavior can be adapteed by first converting string to lowercase of uppercase but was non demanded in the exercise.
            </p>
<ol>
    <li>
        <strong>if input is None: </strong>
        <p>This piece of code is added as the strict typechecking is not working in jupyterbook unlike VS code IDE</p>
    </li>
</ol>
</div>


```python
def is_palindrome(input: str)-> bool:
    if input is None or not isinstance(input ,(str)):
        return False
    else:
        length = len(input)
        if length %2 ==0:
            part1= input[: int(length/2)]
            part2= input[int(length/2):]
            return part1 == part2[None: None: -1]
        else:
            part1= input[: int(length/2)]
            part2= input[int(length/2)+1:]
            return part1 == part2[None: None: -1]
```

<h4>Test case 1</h4>
<p>A palindorme with odd number of characters</p>

```python
print(is_palindrome('helloAolleh'))
```

<h4>Test case 2</h4>
<p>A palindorme with even number of characters</p>

```python
print(is_palindrome('hellolleh'))
```

<h4>Test case 3</h4>
<p> None/Null passed as input</p>

```python
print(is_palindrome(None))
```

<h4>Test case 4</h4>
<p> Invalid data type passed as input</p>

```python
print(is_palindrome(12345))
```

<h4>Test case 5</h4>
<p>Non palindrome string</p>

```python
print(is_palindrome('NotaPalindrome'))
```
