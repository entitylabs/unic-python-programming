## Pandas and DataFrames

```python
import pandas as pd
import ast
import numpy as np

def read_data(path:str)-> pd.DataFrame:
    
    """
    This method reads the data csv file parsing artists as a list
    
    Args:
        path (str): path to file.

    Returns:
        DataFrame: The created data frame.
    """
    
    df = pd.read_csv(path, converters={"artists":lambda item: ast.literal_eval(item)})

    print(df[['name', 'danceability']])
    
    return df

def calculate_statistics(df: pd.DataFrame) -> None:
      
    """
    This method calculates statistics for all numeric fields

     Args:
        df (DataFrame): Dataframe

    Returns:
        None.
    """
    
    print('\nMean:')
    print(df.mean(numeric_only=True))

    print('\nMedian')
    print(df.median(numeric_only=True))

    print('\nMode')
    print(df.mode(numeric_only=True))

def calculate_song_per_artist(df: pd.DataFrame) -> None:

    """
    This method calculates songs per artist

    Args:
        df (DataFrame): Dataframe

    Returns:
        None.
    """
    
    print('\n Songs Per Artist:')
    
    exploded_data_set = df.explode(column="artists")
    exploded_data_set['num_of_songs'] = 1
    
    counted_data_set = exploded_data_set.groupby("artists").sum()
    
    print(counted_data_set[["num_of_songs"]])

def calculate_danceability(df: pd.DataFrame, greater_than:float) -> None:

    """
    This method filters data frame based on danceability value

    Args:
        df (DataFrame): Dataframe
        greater_than (float): The value to filter danceability

    Returns:
        None.
    """
    
    print('\n Filtered Danceability :')
    danceability_gt= df[df['danceability'] > greater_than]
    
    print('\nMean:')
    print(danceability_gt['acousticness'].median())

    print('\nMin:')
    print(danceability_gt['acousticness'].min())
    
    print('\nMax:')
    print(danceability_gt['acousticness'].max())

def classify_danceability(df: pd.DataFrame, v1:float, v2:float) -> None:

    """
    This method classifies danceability based on passed low and high values

    Args:
        df (DataFrame): Dataframe
        v1 (float): Lower range
        v2 (float): Upper range
        
    Returns:
        None
        
    """
    
    print('\n Classified Danceability :')

    validate_range(v1, v2)
    
    danceability_conditions = [
        df["danceability"] < v1,
        ( df["danceability"] >= v1 ) &  ( df["danceability"] < v2 ),
        df["danceability"] > v2]
    
    danceability_levels = ["low_danceability", "medium_danceability", "high_danceability"]
    
    df["danceability_classification"] = np.select(danceability_conditions, danceability_levels, default="low_danceability")
    updated_columns = ['acousticness', 'danceability', 'duration_ms', 'energy', 'explicit', 'instrumentalness', 'liveness', 'loudness', 'popularity', 'speechiness', 'tempo', 'valence']
    df[updated_columns] = df.groupby("danceability_classification")[updated_columns].transform(lambda value: (value- value.mean())/value.std())
    
    print(df[['name', 'danceability', 'danceability_classification']])

def validate_range(v1:float, v2:float)-> None:

    """
    This method validates v1 < v2 else raise an exception

    Args:
        v1 (float): Lower range
        v2 (float): Upper range
        
    Returns:
        None
    """
    
    if v1 < v2:
        return
    
    raise Exception('Invalid range')
```

```python
df = read_data(path="data/data.csv")
```

```python
calculate_statistics(df)
calculate_song_per_artist(df)
calculate_danceability(df, 0.7)
classify_danceability(df, 0.3, 0.7)
```
