<h1>Exercise 1</h1>
<p>
    Create a list of lists where each list has an arbitrary length (use lengths between 1 and 20). The values of the lists can be floats or integers and greater than zero.

<ol>
    <li>
        Truncate the lists that exceed a length of 10
    </li>
    <li>
        Pad the lists with zeros up to length 10
    </li>
</ol>
</p>

<h2>Solution</h2>
<p>
    The problem is solved by first generating a random list of lists. Where each list has random float and int values. Each list has a random size betwwen 1 and 20. Also, the number of such lists created are between 10 and 20
</p>

<ol>
    <li> <strong>generate_random_int_or_float</strong> method randomly generates a float or int value</li>
    <li> <strong>generate_random_list</strong> method generates between 1 and 20 random float and int values to form a list</li>
    <li> <strong>generate_random_list_of_lists</strong> method generates a list of 10 to 20 such random lists</li>
    <li> <strong>generate_truncated_lists</strong> Truncates all such list generated by <strong>generate_random_list_of_lists<strong> having size greater than 10 and pad all lists with zero with size less than 10</li>  
    <li> <strong>create_random_list_and_truncate</strong> This method aggregates the steps of generating list, truncating and printing the results</li>  
</ol>


```python
from typing import List,Union
import random

def generate_random_int_or_float()-> Union[int, float]:
    type=random.randint(0,2)
    return int(random.random()*10000)/100 if type==0 else int(random.random()*100)
    
def generate_random_list():
    list_size = random.randint(1,21)
    return [generate_random_int_or_float() for number in range(list_size)]
    
def generate_random_list_of_lists():
    number_of_lists = random.randint(10,21)
    return [generate_random_list() for number in range(number_of_lists)]
    
def generate_truncated_lists(lists: List[List[Union[float,int]]])-> List[List[Union[float, int]]] :

    truncated_lists: List[List[Union[float, int]]] = [];
    for list in lists:
        if len(list)>=10:
            truncated_list= list[:10]
            truncated_lists.append(truncated_list)
        else:
            padding_list= [0 for num in range(10-len(list))]
            list.extend(padding_list)
            truncated_lists.append(list)
    return truncated_lists


def create_random_list_and_truncate():
    generated_lists = generate_random_list_of_lists();
    print('Randomly generated List of lists:\n',generated_lists,'\n\n');
    truncated_list_of_lists= generate_truncated_lists(generated_lists)
    print('Truncated List of lists:\n',truncated_list_of_lists);    
    
```

<h1>Exercise 2</h1>

<p>Write a function that takes as arguments a list that contain tuples of size 2. Each tuple should have two integers. Sort the list in increasing order by the last element in each tuple.</p>

<ul>
    <li>
        Sample List : [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
    </li>
    <li>
        Expected Result : [(2, 1), (1, 2), (2, 3), (4, 4), (2, 5)]
    </li>
</ul>

<h2>Solution</h2>
<p>
    The problem is solved by using the sort method of list and passing key selector function which selects the second value of tuple as a sort key.
</p>

<ol>
    <li> <strong>get_tuple_key</strong> method selects the second value of the tuple as a list sorting key</li>
    <li> <strong>sort_list_by_tuple_elements</strong> method sorts the list of tuples by second value</li>
</ol>

```python
from typing import Tuple, List

def get_tuple_key(tuple:Tuple[(int,int)])-> int:
    _, key = tuple
    return key

def sort_list_by_tuple_elements(list:List[Tuple[(int,int)]])-> List[Tuple[(int,int)]]:
    list.sort(key=get_tuple_key)
    return list
```

<h4>Test case 1</h4>

```python
print(sort_list_by_tuple_elements([(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]))
```

<h4>Test case 2</h4>

```python
print(sort_list_by_tuple_elements([(100,99),(99,97),(98,95),(300,5)]))
```

<h1>Exercise 3</h1>

<p>Write a function that takes as an argument a square matrix (list of lists) and transposes it.<br />
For example, the following matrix: [[1,2,3],[4,5,6],[6,7,8]] should be transposed to</p>

<p>
   [[1, 4, 6], <br />
   [2, 5, 7], <br />
   [3, 6, 8]]<br />
</p>

<h2>Solution</h2>
<p>
    The soulution first checks the matrix to be transposed is a square method otherwise raises an exception. <br />
    in case the matrix is transposable, the elemets from each array are picked in such a manner that the resulting list of list is transposed.
</p>

<ol>
    <li> <strong>is_matrix_square</strong> method checks if the input matrix is a square matrix, which is also a required pre condition mentioned in the problem statement </li>
    <li> <strong>transpose_matrix</strong> method performs the transpose and retuns the new list of list (the matrix)</li>
    <li> <strong>print_matrix</strong> method prints the list of list into a matrix like structure for easy comprehension</li>
</ol>

```python
from typing import List

def is_matrix_square(matrix:List[List[int]])-> bool:
    column_size=len(matrix)
    is_square_matrix=True
    for row in matrix:
        if not len(row) == column_size:
            return False
    return True

def transpose_matrix(matrix:List[List[int]])->List[List[int]]:
    
    transposed_matrix:List[List[int]]=[]

    if is_matrix_square(matrix):
        
        for index in range(len(matrix)):
            transposed_matrix.append([item[index] for item in matrix])

        return transposed_matrix
    else:
        raise Exception('Matrix is not square')

def print_matrix(matrix:List[List[int]])->List[List[int]]:
    
    print('[')
    
    for row in matrix:
        print(row)
        
    print(']')

def transpose_and_print_matrix(input_matrix:List[List[int]]):
    
    print('Input Matrix:')
    print_matrix(input_matrix)
    
    print('Transposed Matrix:')
    try:
        
        transposed_matrix=transpose_matrix(input_matrix)
        print_matrix(transposed_matrix)
    except:
        print('This operation cannot be performed on a rectangular matrix')
    
    
```

<h4>Test case 1 : A 3x3 square matrix</h4>

```python
transpose_and_print_matrix([[1,2,3],[4,5,6],[6,7,8]])
```

<h4>Test case 2 : A rectangular matrix</h4>

```python
transpose_and_print_matrix([[1,2,3],[3,4]])
```

<h1>Exercise 4</h1>

<p>Write a function that takes two lists as arguments. It creates a dictionary with the first list as keys and the second as values. Then, it reverses the dictionary by creating a new one where values are used as keys and the keys as values. In case where the key in the reversed dictionary exists the function should add this key in a list of duplicates. The function should return the reversed dictionary and the duplicate keys.</p>

<h2>Solution</h2>
<p>
    The soulution first zips the two list and then convert to a dictionary <br />
    Then by iterating the items in a single pass, the keys and values are swapped keeping
    duplicates ina  separate list.
</p>

<ol>
    <li> <strong>create_dictionary_from_lists</strong> method zips the two sepatate list of keys and values and form a dictionary</li>
    <li> <strong>reverse_dictionary</strong> method reverses the dictonary by swapping keys and values and keeping duplicate ones in a separate list and retruning both the new dictionary and the duplicated list in a tuple</li>
    <li> <strong>create_and_reverse_dictionary</strong> method acts as a facade for the above methods</li>
</ol>


```python
from typing import List

def create_dictionary_from_lists(keys_list:List, values_list:List)-> dict:
    dictionary = dict(zip(keys_list, values_list))
    return dictionary;

def reverse_dictionary(dictionary:dict)-> (dict, list):

    reversed_dictionary = dict()
    duplicate_list=[]
    for key,value in dictionary.items():
        if reversed_dictionary.get(value) is None:
            reversed_dictionary[value]=key
        else:
            duplicate_list.append(key);

    return (reversed_dictionary,duplicate_list)


def create_and_reverse_dictionary(keys_list:List, values_list:List)-> (dict, list):
    
    created_dictionary= create_dictionary_from_lists(keys_list, values_list)
    reversed_dictionary, duplicate_list= reverse_dictionary(created_dictionary)
    print('Created Dictionary:', created_dictionary)
    print('Reversed Dictionary:',reversed_dictionary)
    print('Duplicate Keys:',duplicate_list)
    

```
