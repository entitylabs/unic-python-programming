# Cosine Simalirities Utility

This class calculates cosine similarities on a matrix by pair wise comparion of each row with the next row filling in each index. Since this is a utility class with no attributes, it has classmethods only 

* A matrix required is NxM where N<M
  * In case N>M an exception is raised
  * In case of a zero row an exception is raised
*  The cosines is calculated as follows
    1.    The dot product of each vector is calculated ( by calculating dot product of matrix with its transpose creating a nxn matrix
    2.   The normnal of vector each vector is calculated by row wise multiplication of elements, summing them and finding the square root.
    3.   To find norm of other matrix this 1xn generated array is reshaped into 1xn matrix
    4.   The generated norm is transposed into nx1 matrix and multiplied with 1xn norm
    5.   The two norm matrix are multiplis to form a nxn norm matrix
  *   Divide the matrix of dot products by the matrix of norms this will craete cosine of each index or the cosine similarities matrix

```python
""" Module for cosine simlarities functionality """
from typing import NoReturn, Union
import pdb
import numpy as np
import numpy.typing as npt


class CosineSimaliratiesUtility:

    """
    CosineSimaliratiesUtility This class calculates the
    cosine similarites of a NxM matrix
    It throws an exception in case N>M as required in the exercise
    """
    def __init__(self):

        """
        CosineSimaliratiesUtility constructor

        arguments:

        Due to doc string using 'pass' is cossidered redundant by pylint

        """

    @classmethod
    def calculate_cosine_similarities_matrix(cls,
                                             matrix: npt.NDArray[np.float64]
                                             ) -> Union[npt.NDArray[np.float64],
                                                        NoReturn]:
        """
        CosineSimaliratiesUtility calculate_cosine_similarities_matrix method

            arguments:
                matrix: NxM, of floats

            return:
                np.NDArray[np.float64]: A NxN matrix of cos similarities

            Exceptions:
                In case N>M an exception is thrown as it is required
                in the exercise to calculate cos similarities of M>N

        """

        total_rows, total_columns = matrix.shape

        if cls.__is_dimension_mismatch__(matrix):

            error = f"""Cannot find cosine similarites for a matrix size with
            total rows= {total_rows} and total columns = {total_columns}"""

            raise AssertionError(error)

        if cls.__has_zero_row_in_matrix__(matrix):

            raise AssertionError('Cannot find cosine similarites for a matrix with zero row')

        # pdb.set_trace()

        # Calculates the dividend
        vector_dot_products = np.matmul(matrix,np.transpose(matrix))  
        
        matrix.dot(np.transpose(matrix))

        # Calculate the square of each vector component
        vector_square = np.multiply(matrix, matrix)

        # Calculate the norm of each vector
        vector_normals = np.sqrt(np.sum(vector_square, axis=1)).reshape((1,total_rows))

        # Calculate the transpose of each norm
        vector_normals_transposed = np.transpose(vector_normals)
        
        # Calculate norm matrix
        norms_multiplication = np.dot(vector_normals_transposed, vector_normals)
        
        # Calculate cosine matrix by dividing dot product matrix by norm matrix
        cosine_matrix = np.divide(vector_dot_products,norms_multiplication)

        return cosine_matrix


    @classmethod
    def __has_zero_row_in_matrix__(cls, matrix: npt.NDArray[np.float64]) -> np.bool:

        """
        CosineSimaliratiesUtility __has_zero_row_in_matrix__ method. This method ensures 
        in case there is a zero vector, we do not proceed with cosine similarites calculation 
        as this will case divide by zero exceptions

        Logic:
          Check if a row is zero by checking matrix is zero along axis 1 ( rows)
          This generates an array of True and Flase  1xN.
          Now if we multiply this with a transpose of the same matrix of True and Flase
          i.e, Nx1, this will generate a 1x1 matrix and we can safely return the only element 
          at [0][0]

        arguments:
            matrix: npt.NDArray[np.float64]: matrix of float elements

        return:
            bool: The result if all elemts of a row are  Zero

        """

        total_rows, _ = matrix.shape

        has_zero_vector_matrix = np.any(matrix == 0 , axis = 1).reshape(1, total_rows)

        return has_zero_vector_matrix.dot(np.transpose(has_zero_vector_matrix))[0][0]

    @classmethod
    def __is_dimension_mismatch__(cls, matrix: npt.NDArray[np.float64]) -> npt.bool:

        """
        CosineSimaliratiesUtility __is_dimension_mismatch__ method
        This method validates the dimensions of matrix,
        if number of rows are less than equal to the number of columns or matrix is empty

        arguments:
             matrix: npt.NDArray[np.float64]: matrix of float elements

        return:
            bool: is dimension correct

        """

        total_rows, total_columns = matrix.shape
        
        return total_rows > total_columns or total_rows == 0 or total_columns == 0
```
