```python
from typing import List

def insertion_sort(a:List[int])-> List[int]:

    for key_position in range(len(a)):
        
        key = a[key_position]
        
        new_key_position = key_position
        
        for j in reversed(range(0,key_position)):
            
            if(key<a[j]):
                
                a[j+1]=a[j]
                new_key_position = j
                
        a[new_key_position] = key
        
    return a

insertion_sort(a= [10,67,89,1,5,34,23,11,12,99,3,1,1])
```
